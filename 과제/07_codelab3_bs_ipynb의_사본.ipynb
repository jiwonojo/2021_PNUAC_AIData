```python
#라이브러리 불러오기
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs

#파싱
url = 'https://comic.naver.com/webtoon/weekday'
webpage = urlopen(url).read().decode('utf-8')
html = bs(webpage, 'html.parser')

with open('/content/webtoon.csv', 'w') as fp:
  week_list = html.findAll('div',{'class':'col_inner'}) # 월~일 전체 받아오기
  week = ['[월요일]','[화요일]','[수요일]','[목요일]','[금요일]','[토요일]','[일요일]'] # 요일별 정리를 위한 리스트
  line_head = '순위, 제목, 요일, 이미지\n'
  fp.write(line_head)

  for day in range(len(week_list)): #월~일까지 for구문으로 반복
    print(week[day]) #요일 출력
    title = week_list[day].findAll('a',{'class':'title'}) #웹툰 제목이 담겨있는 요소 찾기
    title_list = [t.text for t in title] #텍스트만 리스트에 담기
    li_list = week_list[day].findAll('li') #이미지 부분이 있는 요소 찾기
  
    img_src = [] #이미지 주소를 담기 위한 리스트
    for li in li_list: 
      img = li.find('img') #이미지와 관련된 요소 찾기
      img_src.append(img['src']) #jpg주소부분만 리스트에 담기
    
    index = 1 #웹툰 순위
    for i in title:
      name = i.get_text().replace(',', '')
      data = str(index)+". "+ name #웹툰순위와 제목 합치기
      print(data, img_src[index-1]) #웹춘 순위, 제목, 이미지 링크 출력
      line = f'{str(index)} , {name} , {week[day]} , {img_src[index-1]}\n'
      index += 1
      fp.write(line)
```
